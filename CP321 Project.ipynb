```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv('/Users/ayushchhetri/Downloads/tfwp2022_annual_table_1_e.csv', skiprows=[0,16,17,18,19,20], thousands=',')
display(df)
df.set_index('Province/Territory', inplace=True)

# Transpose the dataframe
df_t = df.T

# Create a line chart
df_t.plot(figsize=(12, 8))

# Add axis labels and title
plt.xlabel('Year')
plt.ylabel('Number of TFW Positions')
plt.title('Trend in the Number of TFW Positions by Province/Territory Over Time')

# Display the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the year you want to analyze
year = '2015','2016', '2017', '2018', '2019', '2020', '2021', '2022'

# Group the dataset by province/territory and sum the TFW positions for the selected year
data = df.groupby('Province/Territory')[year].sum()

# Define the colors for each province/territory
colors = ['red', 'orange', 'yellow', 'green', 'blue', 'purple', 'brown', 'pink', 'gray', 'olive', 'cyan', 'magenta', 'teal']

# Plot the grouped bar chart
fig, ax = plt.subplots(figsize=(12, 8))
data.plot(kind='bar', ax=ax, color=colors)

# Set the chart title and labels
ax.set_title('Total Number of TFW Positions by Province/Territory in {}'.format(year))
ax.set_xlabel('Province/Territory')
ax.set_ylabel('Total Number of TFW Positions')

# Show the chart
plt.show()
```
